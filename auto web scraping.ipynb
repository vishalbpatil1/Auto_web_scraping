```python
url='https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250'
wanted_list = ["Planet Earth II","(2016)","9.5"]
from autoscraper import AutoScraper
scraper = AutoScraper()
# Here we can also pass html content via the html parameter instead of the url (html=html_content)
scraper.build(url,wanted_list)
result=scraper.get_result_similar(url,grouped=True)
```

```python
result.keys()
```

```python
print(result['rule_eqps'][0])
print(result['rule_si3s'][0])
print(result['rule_qhlp'][0])
```

```python
scraper.set_rule_aliases({'rule_eqps':'Rank & Title','rule_si3s':'year','rule_qhlp':'IMDb Rating'})
```

```python
scraper.keep_rules(['rule_eqps','rule_si3s','rule_qhlp'])
```

```python
scraper.save('imdb_tv_show')
```

```python
result=scraper.get_result_similar('https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250',group_by_alias=True)
```
